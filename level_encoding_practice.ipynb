ordinal encoding

```python
df
```

```python
df['income'].unique()
```

```python
income_order = ['low', 'medium', 'high']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enco = OrdinalEncoder(categories=[income_order])
```

```python
enco.fit(df[['income']])
```

```python
final = pd.DataFrame(enco.transform(df[['income']]))
final
```

```python
df['income'] = final
df
```

one hot encoding

```python
df_new = pd.read_csv('homeprices.csv')
df_new
```

```python
dummies = pd.get_dummies(df_new.town)
dummies
```

```python
merged = pd.concat([df_new,dummies],axis='columns')
merged
```

```python
final = merged.drop(['town'],axis='columns')
final
```
